## Step 1 Import Python Modules

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data=pd.read_csv('all_data.csv')
data.head()
```

```python
#renaming column names for better coding 
data = data.rename(columns={'Life expectancy at birth (years)': 'LE'})
data.head()
```

You can start to build graphs from the data by first importing Matplotlib or seaborn and then making some plots!

Some components that you may want to include:

Has life expectancy increased over time in the six nations?
Has GDP increased over time in the six nations?
Is there a correlation between GDP and life expectancy of a country?
What is the average life expectancy in these nations?
What is the distribution of that life expectancy?

```python

f, (ax1, ax2)=plt.subplots(2,1,figsize=(13,13))
sns.set_style("dark")
ax1=sns.lineplot(x='Year', y='LE', data=data, hue='Country', ax=ax1 )
ax1.set_ylabel('Life Expectancy at birth (Years)')
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.1)
ax1.set_title("Life expectancy over time")
ax2=sns.lineplot(x='Year', y='GDP', data=data, hue='Country', ax=ax2 )
ax2.set_ylabel('GDP over time')
ax2.set_ylabel('GDP')
plt.show()
plt.clf()


```

```python
sns.relplot(data=data, x='GDP', y='LE', hue='Country', kind='scatter')
plt.ylabel('Life Expectancy at birth (Years)')
plt.show()
plt.clf()
```
